```python
from NYTAnalysis import *
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)

medsVyears=load_obj('medVyear')
totsVyears=load_obj('totVyear')
years=medsVyears['Years']
meds=medsVyears['MEDs']
tots=totsVyears['TOTs']

# plt.plot(years,meds)
# plt.plot(years,tots)
# plt.show()
plt.close()

fig, ax1 = plt.subplots()
# , figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
# plt.figure(figsize=(8, 6))
fig.set_size_inches(10,6)
t = years
s1 = meds
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('Years')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Median Wordcount', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
s2 = tots
ax2.plot(t, s2, 'k.')
ax2.set_ylabel('Articles', color='k')
ax2.tick_params('y', colors='k')

ax1.set_xlim([1950, 2017])
fig.tight_layout()
plt.show()


```

## Figure 1
### (blue) We show the median word count for a NYTimes article since 1950. We observe since 2009 the median word count has more than doubled from about 350 to more than 700. The last time we saw this behavior was 1970-1980, coinciding with the turmoil of the Johnson and Carter administrations in the lead up to the Reagan administration. 
### (red) We show the number of articles used to calculate the median for each year. It is unlikely that the dataset provided through the API is complete; we show this to demonstrate confidence that the annual median wordcount of an article has converged.
